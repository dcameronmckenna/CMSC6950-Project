```python
import pandas as pd
import matplotlib.pyplot as plt

weather_data = pd.read_csv('en_climate_daily_NL_8403505_2020_P1D.csv', usecols=range(4,31), index_col='Date/Time')

num_cols_weather = ['Total Snow (cm)']

weather_data[num_cols_weather] = weather_data[num_cols_weather].astype(float)

total_snow = weather_data[num_cols_weather]

plt.plot(total_snow[1:30], color = 'g', linestyle = 'dashed', 
         marker = 'o',label = "Weather Data") 
  
plt.xticks(rotation = 25) 
plt.xlabel('Dates') 
plt.ylabel('Total Snow (cm)') 
plt.title('Weather for January 2020', fontsize = 20) 
plt.grid() 
plt.legend() 
plt.show() 
   
```
