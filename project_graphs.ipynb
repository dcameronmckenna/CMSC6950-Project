```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',**{'family':'serif','sans-serif':['Computer Modern Roman']})
rc('text', usetex=True)

weather_data = pd.read_csv('en_climate_daily_NL_8403505_2020_P1D.csv', usecols=range(4,31), index_col='Date/Time')

num_cols_snow = ['Total Snow (cm)']

weather_data[num_cols_snow] = weather_data[num_cols_snow].astype(float)

total_snow = weather_data[num_cols_snow]

axes = plt.subplot()
axes.plot(total_snow[0:31], color = 'cornflowerblue', linestyle = 'dashed', 
         marker = 'o',label = "Snow (cm)") 

axes.set_xticks([0, 14, 30])


plt.xlabel('Dates') 
plt.ylabel('Total Snow (cm)') 
plt.title('Snowfall Amounts for January 2020', fontsize = 20) 
plt.legend() 
plt.savefig("total_snow.pdf", format = "pdf")
plt.show() 
   
```

```python
num_cols_mean_temp = ['Mean Temp (C)']

weather_data[num_cols_mean_temp] = weather_data[num_cols_mean_temp].astype(float)

mean_temp = weather_data[num_cols_mean_temp]

axes = plt.subplot()
axes.plot(mean_temp[0:31], color = 'tab:brown', 
         marker = 'o',label = "Mean Temp (C)") 

axes.set_xticks([0, 14, 30])


plt.xlabel('Dates') 
plt.ylabel('Mean Temp (C)') 
plt.title('Daily Mean Temperature for January 2020', fontsize = 20) 
plt.legend() 
plt.savefig("mean_temp.pdf", format = "pdf")
plt.show() 
   
```

```python
num_cols_max_temp = ['Max Temp (C)']
num_cols_min_temp = ['Min Temp (C)']


weather_data[num_cols_max_temp] = weather_data[num_cols_max_temp].astype(float)
weather_data[num_cols_min_temp] = weather_data[num_cols_min_temp].astype(float)

max_temp = weather_data[num_cols_max_temp]
min_temp = weather_data[num_cols_min_temp]

axes = plt.subplot()
axes.plot(max_temp[0:31], color = 'r', 
         marker = 'o',label = "Max Temp (C)")
axes.plot(min_temp[0:31], color = 'b', 
         marker = 'o',label = "Min Temp (C)") 
axes.axhline(y = 0, color = 'darkorange', label = "Average High", linestyle = "dashed")
axes.axhline(y = -4, color = 'green', label = "Average Low", linestyle = "dashed")

axes.set_xticks([0, 14, 30])


plt.xlabel('Dates') 
plt.ylabel('Temperature (C)') 
plt.title('Daily Temperature for January 2020', fontsize = 20)
plt.legend() 
plt.savefig("maxmin_temp.pdf", format = "pdf")
plt.show() 
```

```python
num_cols_ground = ['Snow on Grnd (cm)']

weather_data[num_cols_ground] = weather_data[num_cols_ground].astype(float)

total_ground = weather_data[num_cols_ground]


bar_data = total_ground[0:31]
axes = bar_data.plot.bar()
axes.set_xticks([0, 14, 30])


plt.xlabel('Dates') 
plt.ylabel('Total Snow on Ground (cm)') 
plt.title('Amount of Snow on Ground for January 2020', fontsize = 20) 
plt.legend() 
plt.savefig("ground_snow.pdf", format = "pdf")
plt.show() 
   
```

```python
num_cols_wind = ['Spd of Max Gust (km/h)']

weather_data[num_cols_wind] = weather_data[num_cols_wind].astype(float)

wind_speeds = weather_data[num_cols_wind]

bar_data_wind = wind_speeds[0:31]
axes = bar_data_wind.plot.bar(color = 'c')
axes.set_xticks([0, 14, 30])

   


plt.xlabel('Dates') 
plt.ylabel('Speed of Max Gust (km/h)') 
plt.title('Daily Max Wind Gusts for January 2020', fontsize = 20) 
plt.legend() 
plt.savefig("wind_gusts.pdf", format = "pdf")
plt.show() 
   
```

```python
num_cols_rain = ['Total Rain (mm)']

weather_data[num_cols_rain] = weather_data[num_cols_rain].astype(float)

total_rain = weather_data[num_cols_rain]

axes = plt.subplot()
axes.plot(total_rain[0:31], color = 'g', linestyle = 'dashed', 
         marker = 'o',label = "Rain (mm)") 

axes.set_xticks([0, 14, 30])


plt.xlabel('Dates') 
plt.ylabel('Total Rain (mm)') 
plt.title('Rainfall Amounts for January 2020', fontsize = 20) 
plt.legend() 
plt.savefig("total_rain.pdf", format = "pdf")
plt.show() 
```
